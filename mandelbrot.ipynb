```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python

a0 = -0.20
a = a0
b0 = 0.25
b = b0

Lx_blue, Ly_blue = [], []

print(f"z0 = {a} + {b} i")

for i in range(1,10):
    a_temporary = a
    a = (a**2 - b**2) + a0
    b = (2 * a_temporary * b) + b0
    print(f"z{i} = {a} + {b} i")
```

```python
def converges(a0, b0):
    a = a0
    b = b0

    for i in range(100):
        a_temporary = a
        a = (a**2 - b**2) + a0
        b = (2 * a_temporary * b) + b0

        if abs(a) > 100 and abs(b) > 100:
            return False
        
    return True
```

```python
def plot():
    Lx_blue, Ly_blue = [], []

    for _ in range(1000000):
        a0 = random.random() * 3 - 2
        b0 = random.random() * 3 - 1.5

        con = converges(a0, b0)

        if con == True:
            Lx_blue.append(a0)
            Ly_blue.append(b0)
        
    
    plt.plot(Lx_blue, Ly_blue, "bo", markersize = 1)
    
    plt.show()
```

```python
plot()
```
