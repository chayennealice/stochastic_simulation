```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python

a0 = -0.20
a = a0
b0 = 0.25
b = b0

Lx_blue, Ly_blue = [], []

print(f"z0 = {a} + {b} i")

for i in range(1,10):
    a_temporary = a
    a = (a**2 - b**2) + a0
    b = (2 * a_temporary * b) + b0
    print(f"z{i} = {a} + {b} i")
```

```python
def converges(a0, b0):
    a = a0
    b = b0

    for i in range(10000):
        a_temporary = a
        a = (a**2 - b**2) + a0
        b = (2 * a_temporary * b) + b0

        if abs(a) > 1000 and abs(b) > 1000:
            return i
        
    return 10000
            
def plot():
    Lx_blue, Ly_blue = [], []
    Lx_yellow, Ly_yellow = [], []
    Lx_green, Ly_green = [], []
    Lx_red, Ly_red = [], []

    for _ in range(1000000):
        a0 = random.random() * 3 - 2
        b0 = random.random() * 3 - 1.5

        con = converges(a0, b0)

        if con == 10000:
            Lx_blue.append(a0)
            Ly_blue.append(b0)
        elif con > 50:
            Lx_green.append(a0)
            Ly_green.append(b0)
        elif con > 25:
            Lx_yellow.append(a0)
            Ly_yellow.append(b0)
        elif con > 10:
            Lx_red.append(a0)
            Ly_red.append(b0)
        
    
    plt.plot(Lx_blue, Ly_blue, "bo", markersize = 1)
    plt.plot(Lx_green, Ly_green, "go", markersize = 1)
    plt.plot(Lx_yellow, Ly_yellow, "yo", markersize = 1)
    plt.plot(Lx_red, Ly_red, "go", markersize = 1)
    
    plt.show()
```

```python
plot()
```
